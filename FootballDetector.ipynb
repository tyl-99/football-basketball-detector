```python
detect_results_dir = 'yolov5/runs/detect/exp3/'

detect_results_files = os.listdir(detect_results_dir)

num_images = len(detect_results_files)
num_cols = 2  
num_rows = (num_images + num_cols - 1) // num_cols  
```

```python
%matplotlib inline

fig, axs = plt.subplots(num_rows, num_cols, figsize=(9, 9))
fig.patch.set_facecolor('black')
axs = axs.flatten()

for i, img_path in enumerate(detect_results_files):
    if i < num_rows * num_cols:
        img = os.path.join(detect_results_dir, img_path)
        
        img = mpimg.imread(img)
        axs[i].imshow(img)
        axs[i].axis('off')

        axs[i].set_facecolor('black')
    else:
        break

for j in range(i + 1, len(axs)):
    axs[j].axis('off')

plt.subplots_adjust(hspace=-0.5)
plt.tight_layout()

plt.show()
```
